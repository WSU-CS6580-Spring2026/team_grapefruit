```python
# Histogram Outliers

# back-link to census for latest information on FIPS

county_xwalk = pd.read_csv(
    "https://www2.census.gov/geo/docs/reference/codes/files/national_county.txt",
    sep=",",
    header=None,
    names=["STUSAB", "STATEFP", "COUNTYFP", "COUNTYNAME", "CLASSFP"],
    dtype=str
)

county_xwalk["fips"] = county_xwalk["STATEFP"] + county_xwalk["COUNTYFP"]


# Identify outlier range

out_low = 6
out_high = 17

tmp = df_A.copy()
tmp["fips"] = tmp["fips"].astype(str)


outliers = (
    tmp.merge(county_xwalk[["fips", "COUNTYNAME", "STUSAB"]], on="fips", how="left")
       .loc[lambda d: ~d["birth_rate_2020_N"].between(out_low, out_high, inclusive="both"),
            ["fips", "COUNTYNAME", "STUSAB", "birth_rate_2020_N"]]
       .sort_values("birth_rate_2020_N")
)

outliers
```

This histogram of 2020 county birth rates reveals an approximately normal distribution with a slight right skew. The distribution appears largely continuous, with no evidence of distinct clustering and only a small number of counties exhibiting markedly different birth rates.

The identified outliers were examined to assess whether these values are plausible and to better understand the types of demographic contexts that may produce extreme observations. While this review is anecdotal rather than research-based, it provides useful intuition for interpreting the data and guiding further exploration.

The three identified low-birth-rate outliers correspond to counties with population structures that limit exposure to childbearing. Charlotte County and Sarasota County in Florida are largely retirement communities, while Hampshire County, Massachusetts, hosts multiple universities, where a large student population may suppress local birth rates.

High-birth-rate outliers appear in counties with distinct demographic or cultural characteristics. Rockland County, New York, which has a large Orthodox and Hasidic Jewish population, and Utah County, Utah, which has a large population of members of The Church of Jesus Christ of Latter-day Saints, suggest that certain religious communities may be associated with higher fertility rates. Onslow County, North Carolina, home to Marine Corps Base Camp Lejeune, indicates that a large population of young military families may also contribute to elevated birth rates. Notably, Utah County is also home to two of the largest universities in the state, suggesting that the relationship between student populations and birth rates may depend on broader cultural and demographic context.

## Scatter Plot

```python
# Scatter Plot


# preliminary definitions

x_col = "birth_rate_2010_N"
y_col = "birth_rate_2020_N"

df_AS = df_A.copy()
df_AS["fips"] = pd.to_numeric(df_AS["fips"], errors="coerce")
df_AS[x_col]  = pd.to_numeric(df_AS[x_col], errors="coerce")
df_AS[y_col]  = pd.to_numeric(df_AS[y_col], errors="coerce")

plot_df = df_AS[["fips", x_col, y_col]].dropna().copy()
plot_df["fips"] = plot_df["fips"].astype(int).astype(str).str.zfill(5)


# add county/state labels
plot_df = plot_df.merge(
    county_xwalk[["fips", "COUNTYNAME", "STUSAB"]],
    on="fips",
    how="left"
)

# 1:1 line

plot_df["above_11"] = plot_df[y_col] > plot_df[x_col]

m, b = np.polyfit(plot_df[x_col], plot_df[y_col], 1)

min_val = min(plot_df[x_col].min(), plot_df[y_col].min())
max_val = max(plot_df[x_col].max(), plot_df[y_col].max())
xs = np.array([min_val, max_val])


# outliers from histogram

label_fips = (
    outliers.assign(fips=pd.to_numeric(outliers["fips"], errors="coerce"))
            .dropna(subset=["fips"])
            ["fips"]
            .astype(int).astype(str).str.zfill(5)
            .unique()
            .tolist()
)


# two new above 1:1 line

non_outlier_above11 = plot_df[
    (~plot_df["fips"].isin(label_fips)) & (plot_df["above_11"])
].copy()

non_outlier_above11["delta_11"] = non_outlier_above11[y_col] - non_outlier_above11[x_col]
top2_above11 = non_outlier_above11.nlargest(2, "delta_11")

# combine outliers with two new
label_df = plot_df[plot_df["fips"].isin(label_fips)].copy()
label_df = pd.concat([label_df, top2_above11], ignore_index=True).drop_duplicates(subset=["fips"])


# color coding

plot_df["is_outlier"] = plot_df["fips"].isin(label_fips)
plot_df["is_low_outlier"]  = plot_df["is_outlier"] & (plot_df[y_col] < out_low)
plot_df["is_high_outlier"] = plot_df["is_outlier"] & (plot_df[y_col] > out_high)

plot_df["category"] = "other"
plot_df.loc[plot_df["is_low_outlier"] & (~plot_df["above_11"]), "category"] = "outlier low below 1:1"
plot_df.loc[plot_df["is_high_outlier"] & ( plot_df["above_11"]), "category"] = "outlier high above 1:1"
plot_df.loc[plot_df["is_high_outlier"] & (~plot_df["above_11"]), "category"] = "outlier high below 1:1"
plot_df.loc[(~plot_df["is_outlier"]) & (plot_df["above_11"]), "category"] = "non-outlier above 1:1"

color_map = {
    "outlier low below 1:1": "orange",
    "outlier high above 1:1": "green",
    "outlier high below 1:1": "yellow",
    "non-outlier above 1:1": "purple",
    "other": "steelblue",
}

plt.style.use("dark_background")


# plot

plt.figure(figsize=(18, 14))

for cat, g in plot_df.groupby("category"):
    plt.scatter(
        g[x_col], g[y_col],
        alpha=0.6,
        color=color_map.get(cat, "gray"),
        label=cat
    )

# 1:1 line (red dashed)
plt.plot([min_val, max_val], [min_val, max_val], linestyle="--", color="red", label="1:1 line")

# OLS fit (blue solid)
plt.plot(xs, m*xs + b, linewidth=2, color="blue", label="OLS fit")

# outlier and new labels
for _, r in label_df.iterrows():
    plt.annotate(
        r["fips"],
        (r[x_col], r[y_col]),
        textcoords="offset points",
        xytext=(6, 6),
        fontsize=8
    )

plt.xlabel("Birth Rate 2010")
plt.ylabel("Birth Rate 2020")
plt.title("County Birth Rates: 2010 vs 2020")
plt.grid(True, alpha=0.3)
plt.legend()

plt.savefig("../results/scatterplotbirthrate20102020.png", dpi=300, bbox_inches="tight")

plt.show()




# data table

emoji_map = {
    "outlier low below 1:1": "🟠",
    "outlier high above 1:1": "🟢",
    "outlier high below 1:1": "🟡",
    "non-outlier above 1:1": "🟣",
    "other": "🔵",
}

table_df = label_df.merge(
    plot_df[["fips", "category"]], on="fips", how="left"
).copy()

table_df["Color"] = table_df["category"].map(emoji_map)
table_df["Notes"] = ""

table_df = table_df[
    ["Color", "fips", "COUNTYNAME", "STUSAB", x_col, y_col, "category"]
].sort_values([y_col, x_col])

table_df
```

This figure plots county-level birth rates in 2010 against those in 2020. The dashed red diagonal line represents a 1:1 reference line, indicating no change in birth rates between the two years. Counties lying above this line buck the overall national trend of declining birth rates and help us identify cases that may shed light on the causes and limits of the generalized decline.

The three low-birth-rate outliers identified in the histogram (orange) and two of the three high-birth-rate outliers (yellow) all lie below the fitted OLS trend line (blue). This suggests that birth rates are falling even more rapidly int he lowest-birth-rate counties and that two of the high-birth-rate outliers are converging toward the broader national trend. In contrast, the remaining high-birth-rate outlier (green) lies above the 1:1 reference line, indicating that--unlike most counties--this county experienced an increase in its birth rate between 2010 and 2020.

In addition to these outliers, a small number of non-outlier counties (purple) also appear above the 1:1 line, indicating modest increases over the decade. Two such counties--Montgomery County, Alabama, and Ocean County, New Jersey--may reflect the influence of military presence and migration-driven population composition, respectively, suggesting avenues for further investigation.

## Heat Map

```python
# Heat map

df_A["female_2529_share"] = df_A["TOT_FEMALE_2529_C"] / df_A["TOT_POP_2529_C"]
df_A["wa_female_2529_share"] = df_A["WA_FEMALE_2529_C"] / df_A["TOT_POP_2529_C"]
df_A["ba_female_2529_share"] = df_A["BA_FEMALE_2529_C"] / df_A["TOT_POP_2529_C"]
df_A["h_female_2529_share"]  = df_A["H_FEMALE_2529_C"]  / df_A["TOT_POP_2529_C"]

df_A.replace([np.inf, -np.inf], np.nan, inplace=True)

cols = [
    "birth_rate_2020_N",
    "birth_rate_2010_N",
    "female_2529_share",
    "wa_female_2529_share",
    "ba_female_2529_share",
    "h_female_2529_share",
    "dem2010_P"
]

corr_df = (
    df_A[cols]
    .apply(pd.to_numeric, errors="coerce")
    .dropna()
)

corr = corr_df.corr()

plt.style.use("default")

plt.figure(figsize=(9, 7))
sns.heatmap(
    corr,
    annot=True,
    fmt=".2f",
    cmap="coolwarm",
    center=0
)
plt.title("Correlation Heatmap: Birth Rates, Demographics, and Voting")
plt.savefig("../results/2529dem2010heatmap.png", dpi=300, bbox_inches="tight")
plt.show()
```

This heat map summarizes pairwise correlations among county-level birth rates, demographic composition, and voting behavior, and highlights several patterns relevant to further investigation. First, the correlation between county birth rates in 2010 and 2020 is approximately 0.90, closely matching the Pearson correlation observed in the earlier scatter plot. This indicates strong persistence in relative birth-rate differences across counties over the decade.

Second, the heat map reveals notable differences in how birth rates are associated with the racial composition of women aged 25 to 29. Counties with higher shares of White women in this age group are negatively correlated with birth rates in both years, while counties with higher shares of Black and Hispanic women show positive correlations. These relationships are broadly stable across 2010 and 2020, suggesting that racial composition is an important cross-sectional correlate of birth rates rather than evidence of sharply changing behavior within groups over time. Such patterns may reflect differences in average age at childbearing, fertility timing, or broader demographic composition across counties.

Finally, the heat map shows little to no correlation between county birth rates and the share of voters supporting the Democratic presidential candidate in 2010 (measured using an aggregate of the 2008 and 2012 elections). This suggests that partisan voting behavior, at least as measured here, is not directly associated with variation in county-level birth rates.
